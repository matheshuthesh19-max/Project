```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print("✅ Libraries Loaded:")
print("  pandas    → Data handling & cleaning")
print("  numpy     → Numerical computation")
print("  matplotlib → Chart visualization")
print("  seaborn   → Advanced visualization")
```

```python
from google.colab import files
uploaded = files.upload()  # 👈 HR-Employee-Attrition.csv select pannu

import io
df = pd.read_csv(io.BytesIO(list(uploaded.values())[0]))

print("✅ Dataset Loaded Successfully!")
print(f"Total Employees : {df.shape[0]}")
print(f"Total Columns   : {df.shape[1]}")
```

```python
print("📋 FIRST 5 ROWS:")
display(df.head())

print("\n📋 ALL COLUMN NAMES:")
print(df.columns.tolist())

print("\n📋 DATASET SHAPE:", df.shape)
```

```python
print("🔍 BEFORE CLEANING:")
print("Missing Values:\n", df.isnull().sum()[df.isnull().sum() > 0])
print("Duplicate Rows:", df.duplicated().sum())
df.fillna(df.mean(numeric_only=True), inplace=True)
df.drop_duplicates(inplace=True)
df.drop(columns=['EmployeeCount', 'Over18', 'StandardHours'], inplace=True)
print("\n✅ AFTER CLEANING:")
print("Missing Values:", df.isnull().sum().sum())
print("Shape:", df.shape)
```

```python
print("📊 STATISTICAL SUMMARY:")
display(df.describe())

print("\n📊 ATTRITION COUNT:")
print(df['Attrition'].value_counts())
print(f"\nAttrition Rate: {round(df['Attrition'].value_counts(normalize=True)['Yes']*100, 2)}%")
```

```python
plt.figure(figsize=(7, 5))
colors = ['#2ecc71', '#e74c3c']
df['Attrition'].value_counts().plot(kind='bar', color=colors, edgecolor='black')
plt.title('Employee Attrition Count', fontsize=14, fontweight='bold')
plt.xlabel('Attrition')
plt.ylabel('Number of Employees')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('attrition_count.png', dpi=150)
plt.show()
print("✅ Chart 1 Saved!")
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Attrition', y='Age', data=df,
            palette={'Yes': '#e74c3c', 'No': '#2ecc71'})
plt.title('Age Distribution by Attrition', fontsize=14, fontweight='bold')
plt.xlabel('Attrition')
plt.ylabel('Age')
plt.tight_layout()
plt.savefig('age_attrition.png', dpi=150)
plt.show()
print("✅ Chart 2 Saved!")
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Department', hue='Attrition', data=df,
              palette={'Yes': '#e74c3c', 'No': '#3498db'})
plt.title('Department-wise Attrition', fontsize=14, fontweight='bold')
plt.xlabel('Department')
plt.ylabel('Count')
plt.legend(title='Attrition')
plt.tight_layout()
plt.savefig('dept_attrition.png', dpi=150)
plt.show()
print("✅ Chart 3 Saved!")
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df,
            palette={'Yes': '#e74c3c', 'No': '#2ecc71'})
plt.title('Monthly Income by Attrition', fontsize=14, fontweight='bold')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.tight_layout()
plt.savefig('income_attrition.png', dpi=150)
plt.show()
print("✅ Chart 4 Saved!")
```

```python
plt.figure(figsize=(14, 10))
numeric_df = df.select_dtypes(include=np.number)
sns.heatmap(numeric_df.corr(), annot=True, fmt='.1f',
            cmap='coolwarm', linewidths=0.5, annot_kws={'size': 7})
plt.title('Correlation Heatmap — All Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('heatmap.png', dpi=150)
plt.show()
print("✅ Chart 5 Saved!")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

overtime_yes = df[df['Attrition'] == 'Yes']['OverTime'].value_counts()
overtime_no  = df[df['Attrition'] == 'No']['OverTime'].value_counts()

axes[0].pie(overtime_yes, labels=overtime_yes.index, autopct='%1.1f%%',
            colors=['#e74c3c', '#f39c12'])
axes[0].set_title('OverTime — Attrition: YES', fontweight='bold')

axes[1].pie(overtime_no, labels=overtime_no.index, autopct='%1.1f%%',
            colors=['#2ecc71', '#3498db'])
axes[1].set_title('OverTime — Attrition: NO', fontweight='bold')

plt.suptitle('OverTime vs Attrition Comparison', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('overtime_attrition.png', dpi=150)
plt.show()
print("✅ Chart 6 Saved!")
```

```python
yes = df[df['Attrition'] == 'Yes']
no  = df[df['Attrition'] == 'No']

print("=" * 60)
print("      📊 HR ATTRITION — FINAL PROJECT INSIGHTS")
print("=" * 60)
print(f"✅ Total Employees Analyzed : {df.shape[0]}")
print(f"✅ Total Features           : {df.shape[1]}")
print(f"✅ Employees Left (Yes)     : {len(yes)}")
print(f"✅ Employees Stayed (No)    : {len(no)}")
rate = round(len(yes)/df.shape[0]*100, 2)
print(f"✅ Attrition Rate           : {rate}%")
print("-" * 60)
print("📌 KEY FINDINGS:")
print(f"  Avg Age (Left)    : {round(yes['Age'].mean(), 1)} yrs")
print(f"  Avg Age (Stayed)  : {round(no['Age'].mean(), 1)} yrs")
print(f"  Avg Income (Left) : ₹{round(yes['MonthlyIncome'].mean(), 0)}")
print(f"  Avg Income (Stay) : ₹{round(no['MonthlyIncome'].mean(), 0)}")
print(f"  OverTime→Left     : {yes['OverTime'].value_counts()['Yes']} employees")
print("-" * 60)
print("🎯 CONCLUSION:")
print("  Low income & overtime employees leave more.")
print("  Sales dept has highest attrition rate.")
print("  Younger employees tend to leave more often.")
print("  Data cleaned, EDA done, 6 visualizations created.")
print("\n✅ Project Completed Successfully!")
```
